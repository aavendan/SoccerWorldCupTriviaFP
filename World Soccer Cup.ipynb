```python
import os
from os.path import *
```

```python
carpetaMundiales = "world-cup"
```

```python
def getFiles(mypath):
    return [f for f in listdir(mypath) if isfile(join(mypath, f))]

def getFolders(mypath):
    return [o for o in os.listdir(mypath) if os.path.isdir(os.path.join(mypath,o))]

def mundiales():
    lstMundiales = []
    for name in getFolders(carpetaMundiales):
        if '--' in name:
            lstMundiales.append(name.replace(carpetaMundiales+"/",""))
    return lstMundiales

def readFile(fileName):
    print("fn ",fileName)
    if os.path.isfile(fileName):
        file = open(fileName, "r", encoding="utf-8")
        lines = file.readlines()
        file.close()
        return lines
    else:
        return list()
    

def mostrarGrupos(anioMundial):
    lstMundiales = mundiales()
    
    for name in lstMundiales:
        if str(anioMundial) in name:
            pathCup = carpetaMundiales+"/"+name
            resultado = [ fileName for fileName in getFiles(pathCup) if  "cup.txt" == fileName ][0]
            pathCupFile = pathCup+"/"+resultado
            content = readFile(pathCupFile)
            
            for linea in content:
                if linea.startswith("Group") and "|" in linea:
                    print(linea.strip())
```

```python
listaMundiales = mundiales()
```

```python
for mundial in listaMundiales:
    nombreMundial = mundial.split("--")
    print(nombreMundial[0], nombreMundial[1].replace("-"," ").title())
    
    anioMundial = int(nombreMundial[0])
    mostrarGrupos(anioMundial)
```
